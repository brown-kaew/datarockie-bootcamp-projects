# Harry potter API

```python
import requests
import pandas as pd
```

```python
# Harry potter api
url = f'https://hp-api.onrender.com/api/characters'

# make get request and parse json
resp = requests.get(url)
characters = resp.json()

names = []
genders = []
houses = []
actors = []

# loop append each character
for character in characters:
  names.append(character['name'])
  genders.append(character["gender"])
  houses.append(character['house'])
  actors.append(character['actor'])

# build dataframe
df = pd.DataFrame({
    "name": names, "gender": genders, "house": houses, "actors": actors
})

df
```
